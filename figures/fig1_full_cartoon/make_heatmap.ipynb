```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('cartoon_heatmap.csv')
df.head()
```

```python
df[df==-3]=-4
df['whole'] = list(np.linspace(4,-4,20))*2
df += np.random.normal(loc=0,scale=1,size=df.values.shape)
```

```python
plt.figure(figsize=[12,6])
sns.heatmap(df,cmap='bwr',vmin=-6,vmax=6)
plt.xticks([])
plt.yticks([])
plt.savefig('cartoon_heatmap.png',bbox_inches='tight',dpi=300)
```
